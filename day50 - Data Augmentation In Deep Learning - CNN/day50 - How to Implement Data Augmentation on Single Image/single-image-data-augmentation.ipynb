```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img("train/cat.jpg", target_size=(200, 200))
```

```python
img
```

```python
type(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2,
)
```

```python
img = image.img_to_array(img)
```

```python
type(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1, 200, 200, 3) # 1 means 1-image
```

```python
i = 0  # start counter at 0

for output in datagen.flow(x=input_batch, batch_size=1, save_to_dir="aug", save_prefix="aug", save_format="jpg"):
    i += 1
    if i == 10:
        break
```
